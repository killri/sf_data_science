## Параметрические тесты

Эти тесты используются для нормально распределенных выборок.
Деляться они на ***коррелционные*** и ***сравнительные***

1. *Коррекляционным параметрическим тестом является* ***тест на корреляцию Пирсона***

Он позволяет понять связаны ли нормально распределенные переменные, без выдвиения гипотез о причинно следственной связи

2. ***t-тест***

Это сравнительный тест. Предполгается, что на входе у нас выборки разделенные по категориальному бинарному параметру (например пол). В результате мы получаем сравнение средних значений этих выборок.
t- тест бывает парный и независимый.
Парный t-тест по сути сравнивает средние значения выбороки и одной сопокупности (например средние баллы М и Ж их одного класса)
Независимый t-тест применяется, если предполагается, что выборки из разных совокупностей. Например средний рост собак и кошек, или баллы учеников разных школ.
t-тест используется для выборок меньше 30ти значений и работает в предположении, что мы не знает дисперсию выборки.

Для выборок больше 30ти обычно испольуют z-тест, но он требует данных о разбросе данных (диспресия или std)

3. ***ANOVA***

Этот сравнительный тест позволяет сравнивать больше 2х выборок, а значит иметь больше 1ой независимых переменных. Сравнивается среднее значение по этим выборкам в предположении нормальности. Например, "какова разница среднего уровня заработка у выпускников 3х разных вузов?"

4. ***MANOVA***

Этот сравнительный тест отличается от ANOVA тем, что позволяет иметь больше одной зависимой переменной. Например "Как влияет вид цветка на длину лепестков, ширину лепестков и длину стебля?"

## Непараметрические тесты

Непараметрические тесты не делают предположений о нормальности данных и не привязаны к типу шкалы измерения признаков. 
Но они тоде бывают ***коррелционные*** и ***сравнительные***

1. ***Тест на корреляцию Спирмана***

Это корреляционный тест, который оценивает *зависимость двух количесвенных величин*. Понимая, что они ненормально распределены. И без выдвижения гипотез о причинно следственной связи.

2.  ***$\chi^2$ критерий*** 

Сравнивает две *категориальных переменных*. В том смысле, что являются ли ожидаемые частоты каждой переменной значимо отличающимися от наблюдаемых частот или нет. 
Например: 10 М и 10 Ж спросили о предпочтениях по Пицце 1 и Пицце 2. Сравниваем если ли различия в выборе по в разрезе М и Ж

3. ***Знаковый ранговый критерий Уилкоксона*** и ***Сумма рангов Уилкоксона***

Критерий Уилкоксона используется для сравнения двух независимых выборок путём анализа их медианных значений.
В каком-то смысле аналог парного t-теста для ненормально распределенных значений.
Знаковый критерий для данных из одной совокупности, а ранги - для данных из разных совокупностей

4. ***Критерий Краскела-Уоллиса***

Сравнивает распределения в в нескольких выборках. Т.е. у нас уже не 2 группы данных, а 3 и больше может быть.

5. ***ANOSIM***

Позволяет брать на входе данные из 3х и более категорий, а данными (зависимой переменной) являются могут 2 и более параметров.
