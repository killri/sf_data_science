Задание 8.1

Найдите минимум функции $2x^2 - 4xy + y^4 + 2$ с помощью градиентного спуска.

```python
def f1(x,y):
    return 2*x**2 - 4*x*y + y**4 +2

def gradf1(x,y):
    dx = 4*x-4*y
    dy = -4*x+4*y**3
    return dx,dy
```

```python
def alg_grad(f,gradf,x0,step,n_iter):
    x = x0
    print(list(range(len(x))))
    vals = []
    coords = []
    x_new = list(range(len(x)))
    i=0
    while True:
        for k in range(len(x)):
            x_new[k] = x[k] - step*gradf(*x)[k]
        if i > n_iter:
            break
        vals.append(f(*x_new))
        coords.append(x_new)
        x = x_new
        i += 1
        print(f"iter={i}; x=({x[0]:.4f}, {x[1]:.4f});"
          f" f(x)={f(*x):.4f}; grad f(x)=({gradf(*x)[0]:.4f}, {gradf(*x)[1]:.4f})")
```

```python
alg_grad(f=f1, gradf=gradf1, x0=[0,1],step=0.1,n_iter=1000)
```

Задание 8.2

Найдите минимум функции $x^3 - 2x^2 + y^2 + z^2 - 2xy + xz - yz + 3z$ с помощью градиентного спуска.

Ответ округлите до целого числа.

```python
def f2(x,y,z):
    return x**3 - 2*x**2 + y**2 + z**2 -2*x*y + x*z - y*z + 3*z

def gradf2(x,y,z):
    dx = 3*x**2 - 4*x - 2*y + z
    dy = 2*y-2*x -z
    dz = 2*z+x-y+3
    return dx,dy,dz
```

```python
alg_grad(f=f2, gradf=gradf2, x0=[0,0,0],step=0.1,n_iter=1000)
```
